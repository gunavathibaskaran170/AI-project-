```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df= pd.read_csv('/content/drive/MyDrive/DATASET /preprocessing_dataset.csv')
df.head()
```

```python
from sklearn.impute import SimpleImputer
im=SimpleImputer(strategy='most_frequent')
imp1=SimpleImputer(strategy='median')
df[['Age','Salary']]=imp1.fit_transform(df[['Age','Salary']])
df[['Gender']]=im.fit_transform(df[['Gender']])
```

```python
from sklearn.preprocessing import LabelEncoder
lc=LabelEncoder()
df['Gender_encoder']=lc.fit_transform(df[['Gender']])
df
```

```python
d2= pd.read_csv('/content/drive/MyDrive/DATASET /preprocessing_dataset.csv')
from sklearn.impute import SimpleImputer
im=SimpleImputer(strategy='most_frequent')
imp1=SimpleImputer(strategy='median')
d2[['Age','Salary']]=imp1.fit_transform(d2[['Age','Salary']])
d2[['Gender']]=im.fit_transform(d2[['Gender']])

from sklearn.preprocessing import OneHotEncoder
oh=OneHotEncoder(sparse_output=False)
d2['Gender']=oh.fit_transform(d2[['Gender']])
d2
```
