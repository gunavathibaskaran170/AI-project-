model works well in the **Float32**

```python
model=SentenceTransformer('all-MiniLM-L6-v2')
doc_embeddings= model.encode(docs).astype("float32") #numerical represntation
```

```python
doc_embeddings.shape
```

L2- euclidean distance

Flat- does not change the word same as it save

adding the embedding with the calulated similarity by L2

```python
dimension = doc_embeddings.shape[1]
index =faiss.IndexFlatL2(dimension)
index.add(doc_embeddings)
```

```python
query ="how the Deep learning related to Machine learning"
query_embedding=model.encode([query]).astype("float32")
top_k=2
distances,indices = index.search(query_embedding ,top_k)
retrieved_chunks =[docs[i] for i in indices[0]]
print("Retrieved Chunks :")
for chunks in retrieved_chunks:
  print("-",chunks)
```
